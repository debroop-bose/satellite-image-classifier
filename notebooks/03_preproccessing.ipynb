```python
# imports

import os
from PIL import Image
import pandas as pd
import numpy as np
from IPython.display import display
from pathlib import Path
import matplotlib as plt
```

```python
# for each image:
    # load, resize, change to array, shape, store temp


images = Path("/Users/ray/Projects/satellite-image-classifier/images").glob("*.jpg")
shapes = []
# iterate through loop, to get hold of each image
# resize them

for path in images:
    img = Image.open(path)
    
    # file path -> var: images
    # photo -> var: img
    # we resize 'img' 
    
    img_resized = img.resize((128, 128))
    
    img_resized = np.asarray(img_resized)
    shapes.append(img_resized.shape)
    
    filename = os.path.basename(path)

    # converting pixel values to float
    
    img_float = (img_resized / 255)
    max = img_float.max()
    min = img_float.min()
    mean = img_float.mean()
    
    print(f"For {filename}: ")
    print("----------------")
    print("Max: ", max)
    print("Min: ", min)
    print("Mean: ", mean)   
    print()

print(shapes)
```
