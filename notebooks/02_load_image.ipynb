```python
from PIL import Image

from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
import os
```

```python
"""
glob helps to find and match filepaths.

for e.g: here, it will look into the folder 'images' and match the photos ending with .jpg % jpeg paths.
"""
images = Path("/Users/ray/Projects/satellite-image-classifier/images").glob("*.jpg")
```

```python
# create a loop
# it will iterate through the array
# take one image (i.e. i)
# load it, then display it small
# compute average RGBs
# store it in a dict and display in a table

red_list = [] 
green_list = [] 
blue_list = [] 
bright_list = [] 
file_list = [] 

for path in images: 
    img = np.asarray(Image.open(path)) # Calculating average 
    avgR = np.mean(img[:, :, 2]) 
    avgG = np.mean(img[:, :, 1]) 
    avgB = np.mean(img[:, :, 0]) 
    avgBright = 0.2126 * avgR + 0.7152 * avgG + 0.0722 * avgB

    red_list.append(avgR) 
    green_list.append(avgG) 
    blue_list.append(avgB) 
    bright_list.append(avgBright) 
    filename = os.path.basename(path) 
    file_list.append(filename) 

    print("FOR FILE: ", path)
    print("========") 

    print("Averaged Red: ", round(avgR, 2)) 
    print("Averaged Green: ", round(avgG, 2)) 
    print("Averaged Blue: ", round(avgB, 2)) 
    print("Averaged Brightness: ", round(avgBright, 2)) 
    print() 

dict = {'File' : file_list, 
        'Red' : red_list, 
        'Green' : green_list, 
        'Blue' : blue_list, 
        'Brightness' : bright_list} 
df = pd.DataFrame(dict) 
display(df)
```
